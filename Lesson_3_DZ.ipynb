*  Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
df['yr_renovated'] = pd.to_datetime(df['yr_renovated'], format='%Y', errors='coerce')
df['yr_built'] = pd.to_datetime(df['yr_built'], format='%Y', errors='coerce')

df['delta_renovated'] = df['yr_renovated'].dt.year - df['yr_built'].dt.year

df['delta_renovated'].fillna(0, inplace=True)

print(df[['yr_built', 'yr_renovated', 'delta_renovated']])
```

* Создайте признаки года продажи, месяца продажи

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d', errors='coerce')

df['год_продажи'] = df['date'].dt.year
df['месяц_продажи'] = df['date'].dt.month

print(df[['date', 'год_продажи', 'месяц_продажи']])
```

*  Удалите признаки date, zipcode, lat, long

```python
df['date'] = 0
df['zipcode'] = 0
df['lat'] = 0
df['long'] = 0
df.head()
```

```python

df.drop(['date', 'zipcode', 'lat', 'long'], axis=1, inplace=True)

print(df.head())
```
